- ### Importando Bibliotecas

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
IFs = df[df["NO_ESCOLA_EDUCACENSO"] == "IF"]
IFs
```

```python
ser_media_part_uf.var()
```

    - **Podemos observa que no que se refere a participação no Enem a maioria dos IFs estão na faixa de 72.8% e 80.5%**

- ### **Será que a taxa de reprovação está relacionado com a participação dos alunos no ENEM?**

```python
part_repro = pd.concat([ser_media_part_uf, ser_media_repro_uf], axis=1)
```

```python
part_repro.plot.scatter(y="NU_TAXA_PARTICIPACAO", x="NU_TAXA_REPROVACAO")
```

```python
part_repro.corr()
```

- ### Observação:
    - Existe uma correlação de -0.16 entre a Taxa de reprovação e a perticipação no Enem

- ### **Será que a taxa de reprovação está relacionado com a nota Geral dos alunos no ENEM?**

```python
ser_media_enem = pd.Series((IFs.loc[:, "NU_MEDIA_CN":"NU_MEDIA_RED"].sum(axis=1)/5), name = "Nota")
```

```python
ser_media_enem_uf = pd.concat([IFs["SG_UF_ESCOLA"], ser_media_enem], axis =1).groupby(["SG_UF_ESCOLA"]).mean()
```

```python
data_media_repro = pd.concat([ser_media_enem_uf, ser_media_repro_uf], axis = 1)
```

```python
data_media_repro.corr()
```

```python
data_media_repro.plot.scatter(x = "Nota", y = "NU_TAXA_REPROVACAO")
```
