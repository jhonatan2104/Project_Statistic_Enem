- # Será que a taxa de reprovação está relacionada com outras variáveis?

Essa pergunta pode ser interpretada de várias formas, incluse podemos encadear várias perguntas. Será que a reprovação dos alunos pode influenciar na participação no Enem? Será que os alunos reprovados chegam a concluir o ensino médio e fazem a prova para uma universidade? 

Podemos concluir isso caparando as taxas de reprovação com a taxa de participação dos IFs no Enem e com a taxa de Abandono

- **Reprovação e Participação no ENEM**

```python
part_repro = pd.concat([ser_media_part_uf, ser_media_repro_uf], axis=1)
```

```python
part_repro.plot.scatter(y="NU_TAXA_PARTICIPACAO", x="NU_TAXA_REPROVACAO")
```

```python
part_repro.corr()
```

- ### Observação:
    - ### Existe uma correlação de -0.16 entre a Taxa de reprovação e a perticipação no Enem

Existe uma mínima correlação negativa entre essas  duas variávais. Dessa forma não podemos afirma muita coisa ecerca da natureza dessas duas variáveis.

- **Reprovação e Abandono**

```python
aban_repro = pd.concat([ser_media_abandono_uf, ser_media_repro_uf], axis=1)
```

```python
aban_repro.plot.scatter(y="NU_TAXA_ABANDONO", x="NU_TAXA_REPROVACAO")
```

```python
aban_repro.corr()
```

- ### Observação:
    - ### Existe uma correlação de 0.56 entre a Taxa de reprovação e de Abandono no IF

Podemos afirma que existe um correlação positiva entre essas duas variávais. Ou seja, de forma cautelosa que existe uma posiblidade de quanto maior é a taxa de reprovação do IF de um estado, maior será a taxa de abandono dos seus alunos. 

- # **Será que a taxa de reprovação está relacionado com a nota Geral dos alunos no ENEM?**

Será que as reprovações são motivas porque é difícil consiliar o técnico com o médio? Ou será que essas reprovações são motivadas pela falta de interesse dos próprios alunos? Essas perguntas possuem respostas bem complexas. Porém, podemos considera-la e analisar alguns dados.
 
Por comparar a taxa de reprovação com as médias das notas do enem por estado podemos obter um indice de correlação.

```python
ser_media_enem = pd.Series((IFs.loc[:, "NU_MEDIA_CN":"NU_MEDIA_RED"].sum(axis=1)/5), name = "Nota")
```

```python
ser_media_enem_uf = pd.concat([IFs["SG_UF_ESCOLA"], ser_media_enem], axis =1).groupby(["SG_UF_ESCOLA"]).mean()
```

```python
data_media_repro = pd.concat([ser_media_enem_uf, ser_media_repro_uf], axis = 1)
```

```python
data_media_repro.corr()
```

```python
data_media_repro.plot.scatter(x = "Nota", y = "NU_TAXA_REPROVACAO")
```

- ### Observação:
    - ### Existe uma correlação de -0.42

Dessa forma, podemos observa uma correlação negativa entre essas duas variáveis. Sá podemos dizer que os estados que possuem as maiores taxas de reprovação possuem um tendência de possuirem menores médias no ENEM
